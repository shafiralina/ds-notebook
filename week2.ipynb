```python
url = 'https://www.olx.co.id/all-results/q-hijab-pashmina/'
req_hijab = requests.get(url)
soup_hijab = BeautifulSoup(req_hijab.text, 'lxml')
```

```python
seluruh_hijab = soup_hijab.find_all('a', {'class':'marginright5 link linkWithHash detailsLink'})
```

```python
seluruh_hijab[0]
```

```python
for hijab in seluruh_hijab:
    print(hijab['href'])
```

```python
for judul in seluruh_hijab:
    print(judul.text.strip())
```

```python
seluruh_hijab
```

```python
seluruh_harga = soup_hijab.find_all('p', {'class':'price large margintop10'})
```

```python
for harga in seluruh_harga:
    print(harga.text.strip()) #strip buat biar rapi, gaada spacing aneh2 dsb
```

```python
seluruh_kota = soup_hijab.find_all('p', {'class':'color-9 lheight14 margintop3'})
```

```python
for kota in seluruh_kota:
    print(kota.find('span').text) 
```

```python
seluruh_tanggal = soup_hijab.find_all('td', {'class':'rel tcenter'})
```

```python
for tanggal in seluruh_tanggal:
    print(tanggal.text.strip())
```

```python
seluruh_gambar = soup_hijab.find_all('img', {'class','fleft'})
```

```python
for gambar in seluruh_gambar:
    print(gambar['src'])
```

```python
hijab = []
for judul, harga, tanggal, kota, gambar in zip(seluruh_hijab, seluruh_harga, seluruh_tanggal, seluruh_kota, seluruh_gambar):
    item = {}
    print(judul.text.strip())
    item['judul'] = judul.text.strip()
    print(harga.text.strip())
    item['harga'] = harga.text.strip()
    print(tanggal.text.strip())
    item['tanggal'] = tanggal.text.strip()
    print(kota.find('span').text.strip())
    item['kota'] = kota.find('span').text.strip()
    print(gambar['src'].strip())
    item['gambar'] = gambar['src'].strip()
    hijab.append(item)
```
