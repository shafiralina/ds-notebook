```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
link = 'https://tekno.kompas.com/search/2019-05-{}'
link4 = ''
```

```python
total_url=[]
hasil=[]
for i in range(4,11):
    total_url.append(link.format(i))
    
for link in total_url:
    req_tekno = requests.get(link)
    soup_tekno = BeautifulSoup(req_tekno.text, 'lxml')
    seluruh_judul_page = soup_tekno.find_all('a', {'class':'article__link'})
    seluruh_tanggal_page = soup_tekno.find_all('div', {'class':'article__date'})
    seluruh_gambar_page = soup_tekno.find_all('div', {'class':'article__asset'})
    
    
    for judul, tanggal, gambar in zip(seluruh_judul_page, seluruh_tanggal_page, seluruh_gambar_page):
        item = {}
        item['judul'] = judul.text.strip()
        item['tanggal'] = tanggal.text.strip()
        item['gambar'] = gambar.a.img['src']
        item['link'] = judul['href']
        req_article = requests.get(judul['href'])
        soup_article = BeautifulSoup(req_article.text, 'lxml')
        seluruh_p_page = soup_article.find('div', {'class':'read__content'})
        item['paragraf'] = seluruh_p_page.text.strip()
        hasil.append(item)
    
    
from pprint import pprint
pprint(hasil)
df = pd.DataFrame(hasil)
df.to_csv('tekno1.csv')
     
```
